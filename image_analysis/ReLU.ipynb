```python
%matplotlib inline
import numpy as np
from matplotlib import pylab
import matplotlib.pyplot as plt
import seaborn
from pond.nn import Relu, ReluExact
from pond.tensor import NativeTensor

              
relu_layer_exact = ReluExact()
relu_layer_approx_3 = Relu()
relu_layer_approx_9 = Relu(order=9)

x = NativeTensor(np.linspace(-1, 1, 100))
y_exact = relu_layer_exact.forward(x)
y_approx_3 = relu_layer_approx_3.forward(x)
y_approx_9 = relu_layer_approx_9.forward(x)

pylab.rcParams['figure.figsize'] = (15, 8)
plt.rcParams.update({'font.size': 22})
plt.tick_params(axis='both', which='major', labelsize=20)
plt.axhline(y=0, c='black')
plt.plot(x.values, y_exact.values, label='Exact', linestyle='--', color='black', linewidth=2.0)
plt.plot(x.values, y_approx_3.values, label='Degree 3', color='blue', linewidth=2.0)
plt.plot(x.values, y_approx_9.values, label='Degree 9', color='red', linewidth=2.0)

plt.legend()
ax = plt.gca()
ax.set_xlabel('x', fontsize=30)
ax.set_ylabel('σ(x)', fontsize=30)
ax.spines['right'].set_position('zero')
ax.spines['left'].set_visible(False) 
ax.spines['top'].set_visible(False) 
ax.spines['bottom'].set_visible(False) 
ax.yaxis.set_label_position('left')
ax.yaxis.set_ticks_position('right')

plt.xticks(np.arange(-1, 2, step=1))
plt.yticks(np.arange(0, 1.5, step=0.5))
fig = plt.gcf() 
```
