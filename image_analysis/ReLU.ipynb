```python
%matplotlib inline
import numpy as np
from matplotlib import pylab
import matplotlib.pyplot as plt
import seaborn
from pond.nn import Relu, ReluExact
from pond.tensor import NativeTensor

              
relu_layer_exact = ReluExact()
relu_layer_approx_3 = Relu()
relu_layer_approx_9 = Relu(order=9)

x = NativeTensor(np.linspace(-1, 1, 100))
y_exact = relu_layer_exact.forward(x)
y_approx_3 = relu_layer_approx_3.forward(x)
y_approx_9 = relu_layer_approx_9.forward(x)

pylab.rcParams['figure.figsize'] = (15, 10)
plt.rcParams.update({'font.size': 22})
plt.plot(x.values, y_exact.values, label='Exact', linestyle='--', color='black')
plt.plot(x.values, y_approx_3.values, label='Order 3', color='blue')
plt.plot(x.values, y_approx_9.values, label='Order 9', color='red')
pylab.title('ReLU approximation')
plt.legend()
ax = plt.gca()
fig = plt.gcf() 
```
